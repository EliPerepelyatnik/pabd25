```python
import pandas as pd
import glob 
import matplotlib.pyplot as plt
import numpy as np
  
raw_data_path = '../data/raw'
file_list = glob.glob(raw_data_path + "/*.csv") 
```

```python
df = pd.read_csv(file_list[0])
```

```python
# заменим district модой (самое часто встречающееся)
most_common_district = df['district'].mode()[0]
df['district'] = df['district'].fillna(most_common_district)
```

```python
# аналогично для house_number, underground, residential_complex
most_common_house_number = df['house_number'].mode()[0]
df['house_number'] = df['house_number'].fillna(most_common_house_number)

most_common_underground = df['underground'].mode()[0]
df['underground'] = df['underground'].fillna(most_common_underground)

most_common_residential_complex = df['residential_complex'].mode()[0]
df['residential_complex'] = df['residential_complex'].fillna(most_common_residential_complex)
```

```python
df.isna().sum()
```

```python
df['price'].hist()
```

```python
x = df['total_meters']
y = df['price']

plt.plot(x, y, 'o')

#obtain m (slope) and b(intercept) of linear regression line
m, b = np.polyfit(x, y, 1)

#add linear regression line to scatterplot 
plt.plot(x, m*x+b)
```

```python
new_df = df[df['total_meters'] < 300]

# Строим график
x = new_df['total_meters']
y = new_df['price']
plt.plot(x, y, 'o')

m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)
plt.show()
```
